**spliting the features and target**

```python
x=heart_data.drop(columns='target',axis=1)
y=heart_data['target']
print(x)
```

```python
print(y)
```

Spliting data into training data and test data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y)
```

```python
print(x.shape,x_test.shape,x_train.shape)
```

Model Training using logistic regression

```python
model=LogisticRegression(max_iter=1000)


#training the logisticRegression modelwith training data
model.fit(x_train,y_train)
```

Model evalution


Accuracy Score

```python
#accuracy  of training data
x_train_prediction=model.predict(x_train)
training_datat_acuuracy_score=accuracy_score(x_train_prediction,y_train)
print("acuuracy of training data",training_datat_acuuracy_score )
```

```python
#accuracy  of test data
x_test_prediction=model.predict(x_test)
training_datat_acuuracy_score=accuracy_score(x_test_prediction,y_test)
print("acuuracy of test data",training_datat_acuuracy_score )
```

Building a predicting system

```python
from scipy.sparse.linalg._dsolve.linsolve import asarray
input_data=(34,0,1,118,210,0,1,192,0,0.7,2,0,2)
#change the input data to a numpy array because it is esy to do reshape a numpy array
input_data_as_numpy_array=np.asarray(input_data)
#reshap a numpy as we are predicting for one data point


input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)


prediction=model.predict(input_data_reshaped)
print(prediction)
```
